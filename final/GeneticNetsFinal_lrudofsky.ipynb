```python
#
# Class node represents a single node of a degree 2 Kaufmann net
# Class net represents a (out)degree 2 Kauffman net with randomly set 
# neighbors and with the neighbor function chosen randomly among all
# 2-input boolean functions
#

from random import *

#
# The sixteen 2-input boolean functions f0, f1, ... , f15
#

def f0(val1,val2):
    return 0

def f1(val1,val2):
    return val1 

def f2(val1,val2):
    return 1-val1 

def f3(val1,val2):
    return val2 

def f4(val1,val2):
    return 1-val2 

def f5(val1,val2):
    if val1+val2 == 0:
        return 0
    else:
        return 1

def f6(val1,val2):
    if val1+val2 == 0:
        return 1
    else:
        return 0

def f7(val1,val2):
    if val1+val2 == 1:
        return 0
    else:
        return 1

def f8(val1,val2):
    if val1+val2 == 1:
        return 1
    else:
        return 0

def f9(val1,val2):
    if val1+val2 == 2:
        return 0
    else:
        return 1
    
def f10(val1, val2):
    if val1+val2 == 2:
        return 1
    else:
        return 0

def f11(val1, val2):
    if val1>val2:
        return 1
    else: 
        return 0

def f12(val1, val2):
    if val1>val2:
        return 0
    else: 
        return 1

def f13(val1, val2):
    if val2>val1:
        return 1
    else: 
        return 0

def f14(val1, val2):
    if val2>val1:
        return 0
    else: 
        return 1

def f15(val1, val2):
    return 1
```

```python
#
# boolEval implements case statement evaluating boolean function of a given
# type with inputs val1 and val2
#

def boolEval(type, val1, val2):
    functions = {0: f0,
               1: f1,
               2: f2,
               3: f3,
               4: f4,
               5: f5,
               6: f6,
               7: f7,
               8: f8,
               9: f9,
               10: f10,
               11: f11,
               12: f12,
               13: f13,
               14: f14,
               15: f15
           }
    f = functions[type] 
    return f(val1,val2)
```

```python
#
# Node class - two neighbors set to empty, type set to 0,
# value randomly set to 0 or 1.
# Note that oldvalue is required in order to correctly update net 
#

class node:

    def __init__(self):
        self.type = 0 
        self.id = 0
        self.neigh1 = None
        self.neigh2 = None
        self.value = choice([0,1])           
        self.oldvalue = self.value

    def setNeigh1(self,aNeigh):
        self.neigh1 = aNeigh

    def setNeigh2(self, aNeigh):
        self.neigh2 = aNeigh

    def evolveValue(self):
        self.value = boolEval(self.type,self.neigh1.oldvalue,self.neigh2.oldvalue)

    def getValue(self):
        return self.value
```

```python
#
# net class - size number of nodes are generated with two random neighbors
# and with random type 
#

class net:

    def __init__(self, size):
        self.size = size
        self.nodes = []
        for i in range(0,size): self.nodes.append(node())
        for i in range(0,size): self.nodes[i].id = i 
        for i in range(0,size): 
            self.nodes[i].setNeigh1(self.nodes[randrange(0,size)])
        for i in range(0,size): 
            self.nodes[i].setNeigh2(self.nodes[randrange(0,size)])
        for i in range(0,size):
            self.nodes[i].type = randrange(0,16)

    def evolveNet(self):
        for i in range(0,self.size):  
            self.nodes[i].evolveValue()
        for i in range(0,self.size):
            self.nodes[i].oldvalue = self.nodes[i].value     

    def printTypes(self):
        for i in range(0,self.size): print(self.nodes[i].type,)
        print()
    
    def listTypes(self):
        types=[]
        for i in range(0,self.size): types+=[self.nodes[i].type]
        return types
        
    def listNeighbors(self):
        neighbors = []
        for i in range(0,self.size): neighbors+=[(self.nodes[i].neigh1.id,self.nodes[i].neigh2.id)]
        return neighbors

    def printNet(self):
        for i in range(0,self.size): print(self.nodes[i].value,)
        print()

    def listvalues(self):
        l = []
        for i in range(0,self.size): l += [self.nodes[i].value]
        return l
```

```python
#
# to run a gene net simulation
# define cycle and generun functions
#

def cycle(newstate, states):
    for i in range(len(states)):
        if states[i] == newstate:
            return len(states) - i,i #returns cyclelen, leaderlen

def generun(size):
    bools = []
    for i in range(0,size): bools.append(randrange(0,16))
    n = net(size)
    current = n.listvalues()
    states = []
    count = 0
    
    while not cycle(current, states):
        n.evolveNet()
        count +=1
        states += [current]
        current = n.listvalues()
    currentcycle = cycle(current, states)
    states += [current]
    print("generun(", size,")")
    print ("cycle found of length:", currentcycle[0])
    print("path to cycle of length:", currentcycle[1])
    for i in states: 
        print(i)
    
    print ("id   neighb   boolean")
    
    neighbors = n.listNeighbors()
    types = n.listTypes()
    for i in range(0,n.size):
        print (i, "  ", neighbors[i], "  ", types[i])


        
generun(10)
```

```python
def generunreturn(n):
    '''Returns (cyclelen,leaderlen)'''
    current = n.listvalues()
    states = []
    count = 0
    
    while not cycle(current, states):
        n.evolveNet()
        count +=1
        states += [current]
        current = n.listvalues()
    currentcycle = cycle(current, states)
    states += [current]
    
    return(currentcycle[0],currentcycle[1])


def boolselect(net, bools, numchanges):
    '''Changes numchanges number of booleans by randomly picking a boolean in the list and changing it to a random new function'''
    
    #Changing numchanges number of boolean values to a random function
    for i in range(0,numchanges):
        bools[randrange(0,len(bools))]=randrange(0,16)
        
    #Changing boolean values of each node in the net
    for i in range(0,net.size): 
        net.nodes[i].type = bools[i]
    return

def net_trials(ntrials, netsize, numchanges):
    trials = []
    bools = []
    for i in range(0,netsize): bools.append(randrange(0,16))
    n = net(netsize)
    
    #Run ntrials number of generuns changing numchanges number of booleans after each run
    for i in range(ntrials):
        trial = generunreturn(n)
        trials += [trial]
        boolselect(n,bools,numchanges)
        
    sumcycles = 0
    sumleaders = 0
    cycles = []
    leaders=[]
    #Used to find the means and medians
    for t in trials: 
        sumcycles+=t[0]
        cycles.append(t[0])
        sumleaders+=t[1]
        leaders.append(t[1])
    
    meancyclelen= (sumcycles/len(trials))
    meanleaderlen=(sumleaders/len(trials))
    cycles.sort()
    leaders.sort()
    mediancyclelen=cycles[int(len(trials)/2)]
    medianleaderlen=leaders[int(len(trials)/2)]
    
    return (meancyclelen,meanleaderlen,mediancyclelen,medianleaderlen)

    
def net_test():
    trials = []
    for i in range(10,50):
        for j in range(1,int(i/10)+1):
            trials+=((i,j),net_trials(100,i,j))
    
    print("NetSize: NumBoolChanges: MCL:    MLL:  MedCL:  MedLL:")
    for i in range(0,len(trials)-1,2):
        print(trials[i][0], "\t", trials[i][1], "\t\t", trials[i+1][0], "\t", trials[i+1][1], "\t",trials[i+1][2], "\t", trials[i+1][3])


net_test()
```
